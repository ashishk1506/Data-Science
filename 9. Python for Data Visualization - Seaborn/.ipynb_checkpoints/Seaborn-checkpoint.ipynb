```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#histogram

sns.distplot(tips['total_bill'],kde=False, bins=30)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
#plot of all numerical values

sns.pairplot(tips, hue='sex', palette="coolwarm")
```

```python
#rugplot

sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
#categorical data
 #categorical colums - non numeric

tips.head()
```

```python
#barplot

sns.barplot(x='sex', y='total_bill', data=tips) #default is mean
```

```python
import numpy as np
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
#countplot

sns.countplot(x='sex', data=tips )
```

```python
#boxplot

sns.boxplot(x='day',y='total_bill', data=tips, hue='smoker')
```

```python
#vailon plot
  #kde distribution + boxplot
    
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
#strip plot

sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', split=True)
```

```python
#swarm plot
  #voilion + strip

sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

```python
#factor plot
 #just pass the kind

sns.factorplot(x='day', y='total_bill', data=tips, kind='bar', estimator=np.sum)
```

```python
sns.countplot(x='day', data=tips)
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(2,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#FUNCTIONAL

plt.plot(x,y)
plt.xlabel('X LABEL')
plt.ylabel('Y LABEL')
plt.title('MY GRAPH')
```

```python
    #CREATING SUBPLOT FUNCTIONAL
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#OBJECT ORIENTED
```

```python

fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X LABEL')
axes.set_title('TITLE')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.5,0.3])

axes1.plot(x,y)
axes1.set_title('Big')
axes2.plot(y,x)
```

```python
      #CREATING SUBPLOT OBJECT ORIENETD
    
fig,axes = plt.subplots(nrows=1, ncols=2) #it automatically adds axes unlike plt.figure()

plt.sca(axes[0])
axes[0].plot(x,y)
axes[0].set_title('Graph 1')

plt.sca(axes[1])
axes[1].plot(y,x)

plt.tight_layout()
```

```python
#figure size and DPI

fig = plt.figure(figsize=(3,1),dpi=100)

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols = 1,figsize=(2,2),dpi=120)

plt.sca(axes[0])
axes[0].plot(x,y)
plt.sca(axes[1])
axes[1].plot(y,x)
plt.tight_layout()
```

```python
#saving the image

fig.savefig('myfig.jpg',dpi=1000)
```

```python
#legends

fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2, label = 'X square')
axes.plot(x,x**3, label='X cube')

axes.legend(loc=0)  #0,1,2,...10 or loc = (0.1,0.1)
```

```python
#apperance

fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x, x**2, color='orange',linewidth = 4, alpha = 0.5, ls = '-.')
axes.plot(x,x**3, color='#000', marker ='o', markersize=10, markerfacecolor='yellow', markeredgewidth=3 )
```

```python
#plot range
 
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**3, color='#000', marker ='o' )
axes.set_xlim([1,4])
axes.set_ylim([20,90])
```
